# SOM - Fraud detection

```python
import pandas as pd
import polars as pl

cca = pd.read_csv('data/som/Credit_Card_Applications.csv')
cca
```

```python
indep = [c for c in cca.columns if c!='Class']
X = cca[indep].values
y = cca.Class.values
X.shape, y.shape
```

## Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
```

```python
X
```

## Train SOM

```python
from minisom import MiniSom
```

```python
som = MiniSom(x=10, y=10, input_len=15)
som.random_weights_init(X)
som.train_random(X, num_iteration=100)
```
